```python
def draw_line(coef,intercept, mi, ma,ax):
    #print(mi,ma)
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[1]*mi - intercept)/coef[0]), mi],[((-coef[1]*ma - intercept)/coef[0]), ma]])
    ax.plot(points[:,0], points[:,1])
```

```python
# here we are creating 2d imbalanced data points 
ratios = [ (100,2),(100, 20), (100, 40), (100, 80)]
C = [0.01,1,100]
fig, ax = plt.subplots(4, 3, figsize=(18, 24))
DataSets = []
for j,i in enumerate(ratios):  
    X_p=np.random.normal(0,0.05,size=(i[0],2))
    X_n=np.random.normal(0.13,0.02,size=(i[1],2))
    y_p=np.array([1]*i[0]).reshape(-1,1)
    y_n=np.array([0]*i[1]).reshape(-1,1)
    X=np.vstack((X_p,X_n))
    y=np.vstack((y_p,y_n))
    DataSets.append([X,y])

for row,data in enumerate(DataSets):
    X,y = data
    for col,c in enumerate(C):
        svm = SVC(C = c,kernel='linear')    
        svm.fit(X,y)
        ax[j,col].set_title(f"D:  C :{c} ,Dataset : {row+1}")
        sv_x = svm.support_vectors_[:,0]
        sv_y = svm.support_vectors_[:,1]
        draw_line(svm.coef_[0],svm.intercept_,np.min(X[:,1]),np.max(X[:,1]),ax[row,col])
        ax[row,col].scatter(X_p[:,0],X_p[:,1])
        ax[row,col].scatter(X_n[:,0],X_n[:,1],color='red')
        ax[row,col].scatter(sv_x,sv_y, color='green', marker='^')
plt.show()
```

# Observations

As we know in SVM that parameter C represents the strength with which we want the SVM model closer to a hard margin and from the experinments it is very clear that when the data is imbalance the value of C need to high to get a good decision boundary which also make a lot of sense intuitively beacuse when the one class is less in number the loss would be already low for and if we make the C low that means we are penealizing it even less
