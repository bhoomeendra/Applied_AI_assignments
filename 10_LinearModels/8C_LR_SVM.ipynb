```python
# you should not import any other packages
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
X= b * np.sin(phi)
Y= a * np.cos(phi)
```

```python
def draw_line(coef,intercept, mi, ma,ax):
    #print(mi,ma)
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[ mi, coef[0]*mi+intercept],[ma , coef[0]*ma+intercept]])
    ax.plot(points[:,0], points[:,1],color='red')
```

```python
alphas=[0.0001, 1, 100] 
outLier = [(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
fig ,ax = plt.subplots(3,5,figsize=(30,18))
for i,alpha in enumerate(alphas):
    train_X = np.copy(X)
    train_y = np.copy(Y)
    for j,pt in enumerate(outLier):
        # plot and then add and you would have to add the same fucntion at the end of the loop
        train_X = np.append(train_X,pt[0])
        train_y = np.append(train_y,pt[1])
        model = SGDRegressor(alpha=alpha,eta0=0.001,learning_rate='constant')
        model.fit(train_X.reshape(-1, 1),train_y)     
        ax[i,j].set_title(f"Alpha : {alpha}, Outlier : {j}")
        ax[i,j].scatter(train_X,train_y)
        # print(model.coef_,model.intercept_)
        draw_line(model.coef_,model.intercept_,np.min(train_X),np.max(train_X),ax[i,j])
plt.show()
```

# Observations

The alpha parameter is for regularization and from the experinments it is clear that when the value of the alpha is high that means more regularization which is resulting in more robust regression(less prone to outliers)

if we have a lot of dimentions then it is better go for l1 regularization as it introduces sparcity when the dimensonality is less its a good idea to go for elasticnet as it is a combination of l1 and l2

```python
fig ,ax = plt.subplots(3,5,figsize=(30,18))
for i,alpha in enumerate(alphas):
    train_X = np.copy(X)
    train_y = np.copy(Y)
    for j,pt in enumerate(outLier):
        # plot and then add and you would have to add the same fucntion at the end of the loop
        train_X = np.append(train_X,pt[0])
        train_y = np.append(train_y,pt[1])
        model = SGDRegressor(alpha=alpha,eta0=0.001,learning_rate='constant',penalty='elasticnet')
        model.fit(train_X.reshape(-1, 1),train_y)     
        ax[i,j].set_title(f"Alpha : {alpha}, Outlier : {j}")
        ax[i,j].scatter(train_X,train_y)
        # print(model.coef_,model.intercept_)
        draw_line(model.coef_,model.intercept_,np.min(train_X),np.max(train_X),ax[i,j])
plt.show()
```
